```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
N0=1
d=1
pi=3.14
inf=20
def mlf(v,beta,t):
    second=0
    for k in range(inf):
        second+=((-d**v*t**v)**k)/math.gamma(v*k+beta)    
    return second
def N_function(t,v):
    first=0
    for n in range(inf):
        second=mlf(v,2*n+2,t)
        first+=((math.gamma(2*n+1)*t**(2*n+1))/(math.gamma(n+1)))*second
    return (N0*2/math.sqrt(pi))*first
```

```python
def plot(v,t):
    alpha_val.add(v)
    l=list(np.arange(1,10))
    N_function_var=[]
    for i in range(0,t):
        N_t=N_function(i,v)
        N_function_var.append(N_t)
        f.write(str(i)+","+str(N_t))
    plt.xlabel('t')
    plt.ylabel("N(t)")
    plt.title("Sol of Eq (2.4) when we fix v: "+str(alpha_val))
    plt.plot(np.arange(t),N_function_var)
    plt.legend(l)
```
